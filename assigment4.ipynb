```python
!pip install transformers accelerate torch --quiet

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
def generate_response(model_name, prompt, max_length=120,
                      do_sample=False, temperature=0.7):

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    inputs = tokenizer(prompt, return_tensors="pt")
    # Move to GPU if available
    if torch.cuda.is_available():
        model = model.to("cuda")
        inputs = {k: v.to("cuda") for k, v in inputs.items()}

    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        do_sample=do_sample,
        temperature=temperature,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
prompt_1 = "List the top three benefits of AI in healthcare."
```

```python
model_small = "EleutherAI/gpt-neo-125M"
model_large = "EleutherAI/gpt-neo-1.3B"

response_small = generate_response(model_small, prompt_1, max_length=80)
response_large = generate_response(model_large, prompt_1, max_length=80)

print("=== GPT-Neo 125M Response ===\n")
print(response_small)
print("\n\n=== GPT-Neo 1.3B Response ===\n")
print(response_large)
```

```python
model_name = "EleutherAI/gpt-neo-125M"

original_prompt = """
Read the following product review and explain how the user's experience with the product
might influence their opinion about recommending it to others. Include aspects of their
satisfaction and any problems they faced.

Product review:
"The product has a sleek design and functions well for basic tasks. However, the battery
drains quickly, which makes it hard to rely on during travel. Overall, it's a decent product,
but the battery issue is a drawback."
"""

response_original = generate_response(model_name, original_prompt, max_length=220)
print("=== Original Prompt Response ===\n")
print(response_original)
```

```python
# question 3
!pip install transformers accelerate torch --quiet

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

def generate_response(model_name, prompt, max_length=180,
                      do_sample=False, temperature=0.7):
    """
    Generate a text response using a causal language model.
    Follows the same structure as the Lab 6 code.
    """
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    if torch.cuda.is_available():
        model = model.to("cuda")

    inputs = tokenizer(prompt, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}

    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        do_sample=do_sample,
        temperature=temperature,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
model_name = "EleutherAI/gpt-neo-1.3B"
```

```python
form_prompt_v1 = """
Create a simple feedback form for a recently launched product.
The form should include:

- Product Name
- Feedback Type (positive / negative / neutral)
- Comments (maximum 50 words).

Use clear placeholders for each field and show the form in plain text so users can fill it in.
"""

response_form_v1 = generate_response(model_name, form_prompt_v1, max_length=200)
print("=== Form Prompt – Version 1 (Initial) ===\n")
print(response_form_v1)
```

```python
form_prompt_v2 = """
You are a system for collecting customer feedback about a newly launched product.
Return a feedback FORM in plain text using exactly the following structure:

Product Name: ________________________________

Feedback Type (circle one): Positive / Negative / Neutral

Comments (maximum 50 words):
______________________________________________
______________________________________________
______________________________________________

Important rules:
- Do NOT fill in any of the fields.
- Do NOT add explanations or extra text.
- Only output the form in this exact layout.
"""

response_form_v2 = generate_response(model_name, form_prompt_v2, max_length=220)
print("=== Form Prompt – Version 2 (Improved) ===\n")
print(response_form_v2)
```
